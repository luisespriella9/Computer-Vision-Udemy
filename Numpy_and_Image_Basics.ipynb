Import Libraries

```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

Opening Images

```
image_path = working_directory+'/00-puppy.jpg'
img = cv2.imread(image_path)
# transform to right color channel
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print("shape of image: ", img.shape)
plt.axis('off')
plt.imshow(img)
plt.show()
```

```
# read image as grayscale
gray_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
# transform to right color channel
gray_img = cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB)
print("shape of image: ", gray_img.shape)
plt.axis('off')
plt.imshow(gray_img, cmap='gray')
plt.show()
```

Resizing Image

```
print("shape of image: ", img.shape)
img_resized = cv2.resize(img, dsize=(1000, 400))
print("shape of image resized: ", img_resized.shape)
plt.axis('off')
plt.imshow(img_resized)
plt.show()
```

Flipping Images

```
horizontal_flip_img = cv2.flip(img, 0)
plt.axis('off')
plt.imshow(horizontal_flip_img)
plt.show()
```

```
vertical_flip_img = cv2.flip(img, 1)
plt.axis('off')
plt.imshow(vertical_flip_img)
plt.show()
```

```
full_flip_img = cv2.flip(img, -1)
plt.axis('off')
plt.imshow(full_flip_img)
plt.show()
```
